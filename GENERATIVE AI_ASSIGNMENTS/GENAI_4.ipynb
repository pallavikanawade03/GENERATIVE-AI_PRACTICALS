ASS-4 Build a simple question-answering chatbot using a pre-trained language model.

Loads a ready-to-use question-answering model

```python
import torch
from transformers import pipeline
```

 Load the pre-trained question-answering model

```python
# Load the pre-trained question-answering model
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
def chat():
    print("Chatbot: Ask me a question or type 'exit' to quit.")

    # Define the context for answering questions
    context = """DistilBERT is a smaller and faster version of BERT, designed for question-answering tasks.
                 It is trained on the SQuAD dataset and can provide answers based on the given context."""

    while True:
        question = input("You: ")
        if question.lower() == "exit":
            print("Chatbot: Goodbye!")
            break

        # Get the model's response
        response = qa_pipeline(question=question, context=context)
        answer = response.get('answer', 'I don’t know the answer to that.')

        print("Chatbot:", answer)
```

```python
# Run the chatbot
chat()
```
