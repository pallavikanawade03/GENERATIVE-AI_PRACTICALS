Uses the transformers library.
Loads a Text Generation Model(distilgpt2)

```python
from transformers import pipeline

generator = pipeline("text-generation", model="distilgpt2")

#Generates Text → Takes a prompt and creates a response.

def generate_response(prompt, max_length=200):
    response = generator(prompt, max_length=max_length, do_sample=True)
    return response[0]["generated_text"]

#Uses Different Prompts

#1) Interview Approach (Answers AI-related questions.)
def interview_approach():
    question = "What is the impact of AI on job markets?"
    q1 = f"Describe AI's impact on job markets."
    q2 = f"What are the positive and negative effects of AI on employment?"
    q3 = f"How should policymakers address AI-driven job displacement?"

    ans1 = generate_response(q1)
    ans2 = generate_response(q2)
    ans3 = generate_response(q3)

    return ans1, ans2, ans3


#2) CoT Prompting (Solves a math problem step by step.)
def cot_prompting():
    prompt = "If a store sells apples at $2 each and oranges at $3 each, and a customer buys 3 apples and 2 oranges, what is the total cost? Think step by step."
    return generate_response(prompt)

#3) ToT Prompting  (Solves a quadratic equation in multiple ways.)
def tot_prompting():
    prompt = "Solve the equation: x^2 - 5x + 6 = 0 by considering different solution paths and selecting the best one."
    return generate_response(prompt)


#4) zero_shot() → Translates without examples.
def zero_shot():
    prompt = "Translate the following English sentence into French: 'The weather is nice today.'"
    return generate_response(prompt)

#5)few_shot() → Translates with a few examples first.
def few_shot():
    prompt = "English: 'Hello, how are you?'\nFrench: 'Bonjour, comment ça va?'\n\nEnglish: 'The weather is nice today.'\nFrench:"
    return generate_response(prompt)

interview_results = interview_approach()
cot_result = cot_prompting()
tot_result = tot_prompting()
zero_shot_result = zero_shot()
few_shot_result = few_shot()


print("Interview Approach Results:", interview_results)
print("\nCoT Result:", cot_result)
print("\nToT Result:", tot_result)
print("\nZero-Shot Result:", zero_shot_result)
print("\nFew-Shot Result:", few_shot_result)
```
